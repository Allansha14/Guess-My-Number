## level 3 : give multiple condition 

1. user shoukd enter the number between 1-100. if it's entering wrong , the program should say ,"invalid number"

2. If the user guessed number>seceret number , then program should say " your number is greater than secret number "

3.  If the user guessed number<seceret number , then program should say " your number is less than secret number "

```python
print(' Welcome to Number Guessing Game !!! ')  
print (' Guess the number between 1-100 ')

seceret_number=53

for i in range (1,8):
    user_input = int(input("please enter your guess"))
    if user_input < 1 or user_input > 100 :
        print('Invalid number ! please enter the number between 1 -100')
    elif user_input > seceret_number :
        print('Your number is greater')
    elif user_input < seceret_number :
        print('Your number is lesser')
    else :
        print ("Well Done! you successfully find the number")
        print(f"Well Done! It took you {i} attempts to guess this secret number ")
        break
        
        
    
```
